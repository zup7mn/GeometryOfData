#### Part A
A Lie Group is a smooth manifold that also forms a group, where the two group operations, multiplication and inverse, are smooth mapping between manifolds. Let's show that
$\text{SL}(2, \mathbb{R})$ is a Lie Group. 

Showing group properites:

1. Say $A, B, C \in \text{SL}(2, \mathbb{R})$. Then by the rules of matrix multiplication
$$
A(BC) = (AB)C
$$

2. The identity matrix $\mathbf{I} \in \text{SL}(2, \mathbb{R})$, since $\det \mathbf{I} = 1$.

3. Say $A \in \text{SL}(2, \mathbb{R})$. Since $\det A \ne 0, A$ is invertible, and by the properties of the determinant:
$$
1 = \det A = \frac{1}{\det A^{-1}} = 1,
$$
so $A^{-1} \in \text{SL}(2, \mathbb{R})$.

The $\text{SL}(2, \mathbb{R})$ is clearly a closed subgroup of $\text{GL}(2)$, so therefore it is a matrix group, and inherits the smooth structure of $\text{GL}(2)$, and thus is also a Lie Group ^[Riemann Geometry Notes, page 18.].

#### Part D



Let's consider the geodesic curves between the gaussians with parameters $(0, 1)$ and $(0, 2)$, and between gaussians with parameters $(0, 4)$ and $(0, 8)$:


$$
\gamma_a(t) = (0, 2^t), \: t \in [0,1]; \; \gamma_b(t) = (0, 2^t), \: t \in [2,3]

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np

class NormalDistribution:
    def __init__(self, mu, sigma):
        self.mu = mu
        self.sigma = sigma

    def pdf(self, x):
        return (1 / (self.sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - self.mu) / self.sigma) ** 2)

g_a_0 = NormalDistribution(0, 1)
g_a_1 = NormalDistribution(0, 2)


fig = go.Figure()

x = np.linspace(-5, 5, 1000)

fig.add_trace(go.Scatter(
    x=x, y=g_a_0.pdf(x),
    mode='lines',
    name = r'$\gamma_a(0)$',
))

fig.add_trace(go.Scatter(
    x=x, y=g_a_1.pdf(x),
    mode='lines',
    name = r'$\gamma_a(1)$',
))

fig.update_yaxes(range=[-.1, 0.6])
#fig.update_layout(width=600, height=600)
fig.update_layout(
    title=r"$\text{PDFs }\gamma_a(0)\text{ and }\gamma_a(1)$"
)
fig.show()
```

```python
g_b_0 = NormalDistribution(0, 4)
g_b_1 = NormalDistribution(0, 8)

fig = go.Figure()

x = np.linspace(-5, 5, 1000)

fig.add_trace(go.Scatter(
    x=x, y=g_b_0.pdf(x),
    mode='lines',
    name = r'$\gamma_b(0)$',
))

fig.add_trace(go.Scatter(
    x=x, y=g_b_1.pdf(x),
    mode='lines',
    name = r'$\gamma_b(1)$',
))

fig.update_yaxes(range=[-.2/6, 0.2])
#fig.update_layout(width=600, height=600)
fig.update_layout(
    title=r"$\text{PDFs }\gamma_b(0)\text{ and }\gamma_b(1)$"
)
fig.show()
```

Before computing the geodesic distances, we first must compute the Riemannian norm of $\gamma'(t)$:

$$
\gamma'(t) = \left( 0, bc e^{ct} \right), \; \| \gamma'(t) \| = \sqrt{\langle  \gamma'(t), \gamma'(t) \rangle _g},
$$
$$
< \gamma'(t), \gamma'(t)>_g = \gamma'(t)^T g \gamma'(t) = \sigma' \frac{2}{\sigma^2} \sigma' = \frac{2}{b^2 e^{ct2}} b^2 c^2 e^{ct2} = 2c^2,
$$
$$
\implies \| \gamma'(t) \|  = \sqrt{2}c.
$$


Our geodesic distances 

$$
L(\gamma_a) = \int_0^1 \| \gamma_a'(t) \| dt = \int_0^1 \sqrt{2}c  dt = \sqrt{2} \ln 2
$$

$$
L(\gamma_b) = \int_2^3 \| \gamma_b'(t) \| dt = \int_2^3 \sqrt{2}c dt = \sqrt{2} \ln 2.
$$


```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=[0, 0], y=[1, 2],
    mode='lines+markers',
    name = r"$\gamma_a$"
))

fig.add_trace(go.Scatter(
    x=[0, 0], y=[4, 8],
    mode='lines+markers',
    name = r"$\gamma_b$"
))

fig.update_yaxes(range=[0, 10])
#fig.update_layout(width=600, height=600)
fig.update_layout(
    title=r"$\text{Curves }\gamma_a\text{ and }\gamma_b\text{ in the Poincaré Half Plane}$"
)
fig.show()
```

## Question 3

First, let's build our binary classifier and gradient descent functions. For logistic regression, our classifier takes the form
$$
p_i = \frac{1}{1+e^{-(\vec{x}_i^T {\theta})}},
$$
where
$$
\vec{x}_i1 = 1,
$$
and p_i is the estimated probability that the $i$-th instance of our data is in class 1. We will often let $z = \vec{x}_i^T {\theta}$.
$$
p = \frac{1}{1+e^{-z}} = \frac{e^z}{e^z + 1} \iff \left( e^z + 1\right) p = e^z 
$$
$$
\iff p e^z + p = e^z \iff p = e^z - e^ z p = e^z (1-p)
$$

$$
\iff \frac{p}{1-p} = e^z.
$$

We will use log likelihood to optimize our parameters:
$$
\ell (\theta) = \sum y_i \log p + (1-y_i) \log \left( 1 - p \right).
$$

We will rearange $\ell (\theta)$ for the $i$-th prediction and do some rearangeing to help us have an easier time finding the gradient. 

\begin{aligned}
\ell_i (\theta) &= y \log p + (1-y) \log \left( 1 - p \right) \\
&= y \log p + \log \left( 1 - p \right) - y \log \left( 1 - p \right) \\
&= \log \left( 1 - p \right) + y \log \left( \frac{p}{ 1 - p} \right) \\
&= \log \left( 1 - \frac{1}{1+e^{-z}}  \right) + y z \\
&= \log \left( \frac{e^{-z}}{1+e^{-z}} \right) + y z \\
&= \log e^{-z} - \log \left( 1 + e^{-z} \right) + y z \\
&= -z - \log \left( 1 + e^{-z} \right) + y z \\
&= -z - \log \left(e^{-z} \left[ e^z + 1 \right] \right) + y z \\
&= -z - \log \left(e^{-z} \right) - \log \left( e^z + 1 \right)+ y z \\
&= -z + z - \log \left( e^z + 1 \right)+ y z \\
&= y z - \log \left( e^z + 1 \right)
\end{aligned}

Remembering that $z = \vec{x}_i^T {\theta}$, we compute the gradient:

\begin{aligned}
\partial_\theta \ell_i (\theta) &= \partial_\theta y_i \vec{x}_i^T {\theta} - \partial_\theta \log \left( e^{\vec{x}_i^T {\theta}} + 1 \right) \\
&= y_i \vec{x_i} - \frac{\vec{x_i} e^{\vec{x}_i^T {\theta}}}{e^{\vec{x}_i^T {\theta}} +1} \\
&= y_i \vec{x_i} - \vec{x_i} p_i \\
&= x_i (y_i - p_i).
\end{aligned}

This is very manageable!


```python
import math
import numpy as np
# for numerical stability, we use a modified log function 
def sigmoid(z):
    return np.where(z >= 0, 1/(1+np.exp(-z)), np.exp(z)/(1+np.exp(z)))

class LogisticRegressor():
    def __init__(self, num_features):
        self.W = np.random.randn(num_features + 1)
    def predict(self, X: np.ndarray):
        x_b = np.insert(X, 0, 1, axis=-1)

        z = x_b @ self.W

        return sigmoid(z)
    
    def loglikelihood(self, X: np.ndarray, y: np.ndarray):
        p = self.predict(X)

        x_b = np.insert(X, 0, 1, axis=-1)

        z = x_b @ self.W

        return np.average(y * z - np.logaddexp(0, z))
    
    def train_one_step(self, X: np.ndarray, y: np.ndarray, lr = 0.01):
        p = self.predict(X)

        x_b = np.insert(X, 0, 1, axis=-1)

        z = x_b @ self.W

        G = (y - p)[:, np.newaxis] * x_b

        grad = np.mean(G, axis=0)

        likelihood = self.loglikelihood(X, y)

        self.W += lr * grad # gradient ascent

        return likelihood, G

    def train_one_epoch_natural(self, X: np.ndarray, y: np.ndarray, lr = 0.01):
        p = self.predict(X)

        x_b = np.insert(X, 0, 1, axis=-1)

        z = x_b @ self.W

        G = (y - p)[:, np.newaxis] * x_b
        grad = np.mean(G, axis=0)
        F = G.T @ G / len(y) + 1e-4 * np.eye(grad.shape[0]) # making sure our matrix stays invertible

        natural_grad = np.linalg.inv(F) @ grad
        likelihood = self.loglikelihood(X, y)
        self.W += natural_grad # * lr

        return likelihood, G
```

```python
import torchvision.datasets as datasets
import matplotlib.pyplot as plt

mnist_train = datasets.MNIST(root="./data", train=True, download=True)

sixes = mnist_train.data[mnist_train.targets == 6]
twos = mnist_train.data[mnist_train.targets == 2]
sixes_flat = sixes.flatten(start_dim=1)
twos_flat = twos.flatten(start_dim=1)

X = np.concatenate([sixes_flat, twos_flat], axis=0).astype(np.float32) / 255.0
y = np.concatenate([np.zeros(len(sixes_flat)), np.ones(len(twos_flat))], axis=0)

lr_normal = LogisticRegressor(num_features=28*28)
lr_natural = LogisticRegressor(num_features=28*28)

lr_natural.W = lr_normal.W.copy()

likelihoods_normal = []
likelihoods_natural = []
grads = []
```

```python
for i in range(100):
    ll_normal, _ = lr_normal.train_one_step(X, y, lr=0.1)
    ll_natural, _ = lr_natural.train_one_epoch_natural(X, y)
    likelihoods_normal.append(ll_normal)
    likelihoods_natural.append(ll_natural)
```

```python
import plotly.graph_objects as go
fig = go.Figure()

fig.add_trace(go.Scatter(
    y=likelihoods_normal,
    mode='lines',
    name='Normal Gradient'
))

fig.add_trace(go.Scatter(
    y=likelihoods_natural,
    mode='lines',
    name='Natural Gradient'
))

fig.update_layout(
    title='Log Likelihoood',
    xaxis_title='Epoch',
    yaxis_title='Log-Likelihood'
)

fig.show()
```

Incredible! Using the natural gradient makes training converge much faster than without the normal gradient. Running them with different random initializations shows that the natural gradient always converges to close to zero after only 10-25 iterations, while it takes the normal gradient much longer to lower before leveling out noticably below 0.
