```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = 3
samples = 100000
rn_g = np.random.default_rng(seed=42)
w = np.zeros(shape=(d))
```

```python
X = rn_g.multivariate_normal(np.zeros(d), np.identity(d), samples)
```

```python
norms = []
for i in range(1000000):
    X = rn_g.normal(size=(d))
    X = X / np.linalg.norm(X)  # Normalize to lie on the unit sphere in R^d
    norms.append(np.linalg.norm(X[:d-1]))


np.array(norms).std()
```

```python
plt.hist(norms, bins=30)
```
